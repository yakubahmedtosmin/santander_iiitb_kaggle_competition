```python
import numpy as np
import pandas as pd
```

```python
feature_cols = ['ind_empleado', 'pais_residencia', 'sexo',
       'age', 'fecha_alta', 'ind_nuevo', 'antiguedad', 'indrel',
       'ult_fec_cli_1t', 'indrel_1mes', 'tiprel_1mes', 'indresi', 'indext',
       'conyuemp', 'canal_entrada', 'indfall', 'tipodom', 'cod_prov',
       'nomprov', 'ind_actividad_cliente', 'renta', 'segmento']
```

```python
train = pd.read_csv("../input/santander-pr/train.csv", usecols = feature_cols)
test = pd.read_csv('../input/santander-pr/test.csv', usecols = feature_cols)
```

```python
train.shape, test.shape
```

```python
target_cols = ['ind_ahor_fin_ult1','ind_aval_fin_ult1','ind_cco_fin_ult1',
               'ind_cder_fin_ult1','ind_cno_fin_ult1','ind_ctju_fin_ult1',
               'ind_ctma_fin_ult1','ind_ctop_fin_ult1','ind_ctpp_fin_ult1',
               'ind_deco_fin_ult1','ind_deme_fin_ult1','ind_dela_fin_ult1',
               'ind_ecue_fin_ult1','ind_fond_fin_ult1','ind_hip_fin_ult1',
               'ind_plan_fin_ult1','ind_pres_fin_ult1','ind_reca_fin_ult1',
               'ind_tjcr_fin_ult1','ind_valo_fin_ult1','ind_viv_fin_ult1',
               'ind_nomina_ult1','ind_nom_pens_ult1','ind_recibo_ult1']
```

```python
print(train.columns[train.isnull().sum()>0])
print(test.columns[test.isnull().sum()>0]) 
```

```python
train["renta"]   = pd.to_numeric(train["renta"], errors="coerce")
test["renta"]   = pd.to_numeric(test["renta"], errors="coerce")
train.renta.fillna(train["renta"].mode(), inplace=True)
test.renta.fillna(test["renta"].mode(), inplace=True)
train.cod_prov.fillna("28.0", inplace = True)
test.cod_prov.fillna("28.0", inplace = True)
train.nomprov.fillna("MDRID",inplace = True)
test.nomprov.fillna("MADRID",inplace = True)
train["age"]   = pd.to_numeric(train["age"], errors="coerce")
test["age"]   = pd.to_numeric(test["age"], errors="coerce")
train['ind_empleado'].fillna('N',inplace=True)
test['ind_empleado'].fillna('N',inplace=True)
train['pais_residencia'].fillna('ES',inplace=True)
test['pais_residencia'].fillna('ES',inplace=True)
train['ind_nuevo'].fillna(0.0,inplace=True)
test['ind_nuevo'].fillna(0.0,inplace=True)
train['sexo'].fillna('V',inplace=True)
test['sexo'].fillna('V',inplace=True)
train['indrel'].fillna(1.0,inplace=True)
test['indrel'].fillna(1.0,inplace=True)
train['renta'].fillna((train['renta'].median()),inplace=True)
test['renta'].fillna((test['renta'].median()),inplace=True)
train['age'].fillna((train['age'].median()),inplace=True)
test['age'].fillna((test['age'].median()),inplace=True)
train.segmento.fillna('02 - PARTICULARES',inplace=True)
test.segmento.fillna('02 - PARTICULARES',inplace=True)
train.indrel_1mes.fillna(0, inplace = True)
test.indrel_1mes.fillna(0, inplace = True)
train.tiprel_1mes.fillna('I', inplace = True)
test.tiprel_1mes.fillna('I', inplace = True)
train.indext.fillna('S', inplace = True)
train.conyuemp.fillna('N', inplace = True)
test.conyuemp.fillna('N', inplace = True)
train.canal_entrada.fillna('KHE', inplace = True)
test.canal_entrada.fillna('KHE', inplace = True)
train.indfall.fillna('N', inplace = True)
train.ind_actividad_cliente.fillna(0.0, inplace = True)
```

```python
from sklearn import preprocessing, ensemble
for ind,col in enumerate(train.columns):
    print(col)
    if train[col].dtype == 'object':
        le = preprocessing.LabelEncoder()
        le.fit(list(train[col].values))
        temp_train_X = le.transform(list(train[col].values)).reshape(-1,1)[st:,:]
    else:
        temp_train_X = np.array(train[col]).reshape(-1,1)[st:,:]
    if ind == 0:
        train_X = temp_train_X.copy()
    else:
        train_X = np.hstack([train_X, temp_train_X])
```

```python
## reshape and creating the traget data set: 
train_Y = pd.read_csv("../input/santander-pr/train.csv", usecols=['ncodpers']+target_cols)
drop_duplicate = train_Y.drop_duplicates('ncodpers', keep='last')
train_Y = np.array(train_Y.fillna(0)).astype('int')[st:,1:]
```

```python
train_X.shape, train_Y.shape
```

```python
rfc = ensemble.RandomForestClassifier(n_estimators=29,
                 max_depth=55, min_samples_leaf=15, n_jobs=5, random_state=2016)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import KFold
from sklearn.metrics import auc, roc_auc_score
rfe = RFECV(estimator=rfc, step=10, cv=KFold(n_splits=5, shuffle=False),
            scoring='accuracy', verbose=2) 
```
